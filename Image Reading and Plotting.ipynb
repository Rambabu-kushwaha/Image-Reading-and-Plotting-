```python
import cv2
import matplotlib.pyplot as plt

# Upload the image
from google.colab import files
uploaded = files.upload()

# Assuming the uploaded image is called 'peakpx (1).jpg'
image_path = 'peakpx (1).jpg'

# Read the color image in BGR format
color_image_bgr = cv2.imread(image_path)
if color_image_bgr is None:
    print("Error: Unable to load image.")
else:
    # Convert BGR to RGB
    color_image_rgb = cv2.cvtColor(color_image_bgr, cv2.COLOR_BGR2RGB)

    # Read the grayscale image
    gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Plot the images
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(color_image_rgb)
    plt.title('Color Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
